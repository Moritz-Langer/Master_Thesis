```python
import torch
import torchgeo
from torchgeo.datasets import (
    IntersectionDataset,
    RasterDataset,
    BoundingBox,
    stack_samples,
)
from torchgeo.samplers import RandomGeoSampler, GridGeoSampler
from torch.utils.data import DataLoader
import segmentation_models_pytorch as smp
from torchmetrics.classification import MulticlassJaccardIndex, MulticlassAccuracy
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import os
from torchmetrics.classification import MulticlassConfusionMatrix

# --- 1. Configuration & Setup ---
DRIVE_PATH = r"C:\Users\moritz.langer\Documents\Masterarbeit_Orthophoto_Segmentierung_Klassifikation"
NIR_IMG_PATH = r"C:\Users\moritz.langer\Downloads\2024350_Mosaik_NIR.tif"
RGB_IMG_PATH = r"C:\Users\moritz.langer\Downloads\2024350_Mosaik_RGB.tif"
MASK_IMG_PATH = r"C:\Users\moritz.langer\Documents\Masterarbeit_Orthophoto_Segmentierung_Klassifikation\Deep_Learning\UNet\final_lulc_train_remapped_2.tif"
MODEL_PATH = r"C:\Users\moritz.langer\Documents\Masterarbeit_Orthophoto_Segmentierung_Klassifikation\Deep_Learning\SegFormer\best_model_24.pth"

# Model and Data Configuration (match training)
NUM_CLASSES = 40
IGNORE_INDEX = -1 # The value that the loss function and metrics will ignore
PATCH_SIZE = 512
BATCH_SIZE = 4
num_samples = 1500
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
NORMALIZATION_FACTOR = 255.0

print(f"Using device: {DEVICE}")
print(f"Loading model from: {MODEL_PATH}")

# --- 2. Data Loading with Correct Transformations ---

# Datasets
rgb_data = RasterDataset(paths=RGB_IMG_PATH, crs=31256)
nir_data = RasterDataset(paths=NIR_IMG_PATH, crs=31256)
# Apply the remapping transform to the mask dataset
mask_dataset = RasterDataset(paths=MASK_IMG_PATH, crs=31256)
mask_dataset.is_image = False

rgb_nir_data = IntersectionDataset(rgb_data, nir_data)
dataset = IntersectionDataset(rgb_nir_data, mask_dataset)

# ROI definition
train_roi = BoundingBox(minx=-101193.4376, maxx=-47909.23964, miny=172175.6434, maxy=225951.9085, mint=0, maxt=0)
val_roi = BoundingBox(minx=-47909.23964, maxx=-36491.19722, miny=172175.6434, maxy=225951.9085, mint=0, maxt=0)
test_roi = BoundingBox(minx=-36491.19722, maxx=-25073.1548, miny=172175.6434, maxy=225951.9085, mint=0, maxt=0)

# Test DataLoader
# test_sampler = RandomGeoSampler(dataset, size=PATCH_SIZE, length=num_samples, roi=test_roi) # Using 100 samples for faster eval
test_sampler = GridGeoSampler(dataset, size=PATCH_SIZE, stride=PATCH_SIZE, roi=test_roi)
test_dataloader = DataLoader(dataset, sampler=test_sampler, batch_size=BATCH_SIZE, collate_fn=stack_samples)

# --- 3. Model and Metrics Initialization ---

# Model
model = smp.Segformer(
    encoder_name="mit_b3",
    encoder_weights="imagenet",
    in_channels=4,          # RGB + NIR
    classes=NUM_CLASSES, # Model outputs 39 channels, corresponding to classes 0-38
)

model.load_state_dict(torch.load(MODEL_PATH, map_location=DEVICE))
model.to(DEVICE)
model.eval()

# Metrics (ensuring no-data is ignored)
# Overall mIoU
test_miou = MulticlassJaccardIndex(num_classes=NUM_CLASSES, ignore_index=IGNORE_INDEX).to(DEVICE)
# Per-class IoU
test_per_class_iou = MulticlassJaccardIndex(num_classes=NUM_CLASSES, average='none', ignore_index=IGNORE_INDEX).to(DEVICE)
# Overall Accuracy
test_accuracy = MulticlassAccuracy(num_classes=NUM_CLASSES, ignore_index=IGNORE_INDEX).to(DEVICE)

test_conmat = MulticlassConfusionMatrix(num_classes = NUM_CLASSES, ignore_index=IGNORE_INDEX,normalize="true" ).to(DEVICE) 
# --- 4. Inference and Metric Calculation ---
images_to_viz, truths_to_viz, preds_to_viz = [], [], []

with torch.no_grad():
    for batch in tqdm(test_dataloader, desc="Evaluating on Test Set"):
        images = batch['image'].to(DEVICE).float()
        masks = batch['mask'].to(DEVICE)

        # Squeeze mask dimension
        if masks.dim() == 4 and masks.shape[1] == 1:
            masks = masks.squeeze(1)

        images_normalized = images / NORMALIZATION_FACTOR

        # Get model predictions
        outputs = model(images_normalized)
        preds = torch.argmax(outputs, dim=1)

        mask_in = masks != -1

        
        
        # Only proceed if there are any valid pixels in this batch
        if mask_in.any():
            masks_for_metric = masks.clone()

            # Filter both preds and masks to only the valid pixels
            valid_preds = preds[mask_in]
            valid_masks = masks_for_metric[mask_in]

            # Update metrics with the filtered (and non-empty) tensors
            test_miou.update(valid_preds, valid_masks)
            test_per_class_iou.update(valid_preds, valid_masks)
            test_accuracy.update(valid_preds, valid_masks)
            test_conmat.update(valid_preds, valid_masks)
        

        # Store a few batches for visualization
        if len(images_to_viz) < 5:
            images_to_viz.extend(images.cpu())
            truths_to_viz.extend(masks.cpu())
            preds_to_viz.extend(preds.cpu())

# --- 5. Display Evaluation Results ---
print("\n--- Final Test Results ---")
print(f"Overall Mean IoU (mIoU): {test_miou.compute().item():.4f}")
print(f"Overall Accuracy:        {test_accuracy.compute().item():.4f}")

print("\n--- Per-Class IoU ---")
per_class_iou = test_per_class_iou.compute()
for i, iou in enumerate(per_class_iou):

    print(f"  Class {class_data[str(i)]["name"]}: {iou.item():.4f}")

# --- 6. Visualization ---
print("\n--- Visualizing Predictions ---")

# Define a colormap
colors = plt.cm.get_cmap('viridis', NUM_CLASSES)
custom_cmap_list = colors(np.arange(NUM_CLASSES))
# Use black for the ignore index
custom_cmap_list = np.vstack((custom_cmap_list, [0, 0, 0, 1]))
custom_cmap = plt.matplotlib.colors.ListedColormap(custom_cmap_list)

def plot_sample(image, ground_truth, prediction):
    # Clip image for visualization and select RGB channels
    img_rgb = np.clip(image.numpy().transpose(1, 2, 0)[:, :, :3] / (NORMALIZATION_FACTOR), 0, 1)

    gt_np = ground_truth.numpy()
    pred_np = prediction.numpy()

    # Mask out the prediction where the ground truth is ignored for a fair comparison
    pred_masked = np.where(gt_np == IGNORE_INDEX, IGNORE_INDEX, pred_np)

    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    
    # Plot Image
    axes[0].imshow(img_rgb)
    axes[0].set_title("Input Image (RGB)")
    axes[0].axis('off')

    # Plot Ground Truth
    # Map ignore_index (-1) to the last color in the colormap (NUM_CLASSES)
    gt_viz = np.where(gt_np == IGNORE_INDEX, NUM_CLASSES, gt_np)
    axes[1].imshow(gt_viz, cmap=custom_cmap, vmin=0, vmax=NUM_CLASSES)
    axes[1].set_title("Ground Truth Mask")
    axes[1].axis('off')

    # Plot Prediction
    axes[2].imshow(pred_np, cmap=custom_cmap, vmin=0, vmax=NUM_CLASSES)
    axes[2].set_title("Model Prediction")
    axes[2].axis('off')

    plt.tight_layout()
    plt.show()

    # Print unique values and counts
    gt_unique, gt_counts = np.unique(gt_np, return_counts=True)
    pred_unique, pred_counts = np.unique(pred_np, return_counts=True)
    print("Ground Truth Unique Values & Counts:")
    # Map class indices back for readability
    gt_dict = {f"Class {k}": v for k, v in zip(gt_unique, gt_counts)}
    print(gt_dict)

    print("\nPrediction Unique Values & Counts (in ignored areas):")
    pred_dict = {f"Class {k}": v for k, v in zip(pred_unique, pred_counts)}

    print(pred_dict)
    print("-" * 50)


# Visualize the first 5 samples
for i in range(min(10, len(images_to_viz))):
    plot_sample(images_to_viz[i], truths_to_viz[i], preds_to_viz[i])

# fig_, ax_ = test_conmat.plot()
```

```python
import pandas as pd
import datetime

# --- 7. Exporting Results to CSV ---

print("\n--- Exporting results to CSV ---")

# --- A. Prepare Overall Metrics ---
overall_metrics = {
    'model_name': 'SegFormer',
    'encoder': 'mit_b3',
    'model_path': MODEL_PATH,
    'mIoU': test_miou.compute().item(),
    'overall_accuracy': test_accuracy.compute().item(),
    'timestamp': datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
}

# Create a DataFrame for the overall results
df_overall = pd.DataFrame([overall_metrics])

# Define the output file path
output_file_overall = os.path.join(DRIVE_PATH, "evaluation_summary.csv")

# Save the overall results
# Check if the file exists to append or write a new header
if os.path.exists(output_file_overall):
    df_overall.to_csv(output_file_overall, mode='a', header=False, index=False)
    print(f"Appended overall results to {output_file_overall}")
else:
    df_overall.to_csv(output_file_overall, mode='w', header=True, index=False)
    print(f"Created new overall results file: {output_file_overall}")


# --- B. Prepare Per-Class Metrics ---


try:
    class_names = [class_data[str(i)]["name"] for i in range(NUM_CLASSES)]
except NameError:
    print("Warning: 'class_data' not found. Using generic class names.")
    class_names = [f"Class_{i}" for i in range(NUM_CLASSES)]


per_class_iou_values = test_per_class_iou.compute().cpu().numpy()

# Create a DataFrame for per-class IoU
df_per_class = pd.DataFrame({
    'class_id': list(range(NUM_CLASSES)),
    'class_name': class_names,
    'iou': per_class_iou_values
})

# Define a unique filename for this run's per-class results
timestamp_str = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
output_file_per_class = os.path.join(DRIVE_PATH, f"per_class_iou_SegFormer_mit_b3_{timestamp_str}.csv")

# Save the per-class results to a new file
df_per_class.to_csv(output_file_per_class, index=False)
print(f"Saved per-class IoU results to: {output_file_per_class}")


# --- C. Prepare and Export the Confusion Matrix ---
conmat_values = test_conmat.compute().cpu().numpy()

# Create a DataFrame from the confusion matrix
df_conmat = pd.DataFrame(conmat_values,
                         index=class_names,    # Rows are True Labels
                         columns=class_names)  # Columns are Predicted Labels

# Define a unique filename for this run's confusion matrix
output_file_conmat = os.path.join(DRIVE_PATH, f"confusion_matrix_SegFormer_mit_b3_{timestamp_str}.csv")
df_conmat.to_csv(output_file_conmat, index=True) # index=True to save the row labels
print(f"Saved confusion matrix to: {output_file_conmat}")
```
